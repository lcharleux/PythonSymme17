Une équation qui relie $\alpha$ à $x$:

$$
\alpha = \int_0^\infty x dx
$$

```python
def fonction(x):
    """ 
    Ma fonction mathématique
    """
    return np.sin(x**2) * x**2

   
# Commentaire
fonction(5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
#x = np.array([1,2,3])
x = np.linspace(0., 10., 1000) # Grille de points de 0 à 10 avec 1000 pts
y = fonction(x)

fig = plt.figure()
plt.plot(x, y)
plt.grid()
plt.xlabel("Axe $x$")
plt.ylabel("Axe $y$")
plt.show()
```
