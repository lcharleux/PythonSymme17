
# Création d'un maillage simple et gestion de propriétés

## Création d'un maillage

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
import matplotlib as mpl
%matplotlib nbagg
"""
nodes = np.random.rand(100, 2)
nodes[0] = [0,0]
"""
x , y = np.linspace(0., 1, 10), np.linspace(0., 1, 10)
X, Y = np.meshgrid(x, y)
nodes = np.array([X.flatten(), Y.flatten()]).transpose()
print nodes.shape
conn = Delaunay(nodes).simplices.copy()
nodes[:,0] *= 1. + nodes[:,1]
elements = []
for c in conn:
    element = mpl.patches.Polygon(nodes[c])
    elements.append(element)

```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(nodes[:,0], nodes[:,1], "o", label = "nodes")

mesh = mpl.collections.PatchCollection(elements, 
                                       cmap=mpl.cm.inferno, 
                                       alpha=.5)

mesh.set_array(np.random.rand(5))
ax.add_collection(mesh)
cbar = plt.colorbar(mesh)
cbar.set_label(r"Champ $\alpha$")
plt.axis("off")
plt.grid()
plt.legend()
plt.savefig("mesh.pdf")
plt.show()
```
