## Création des données



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib nbagg
```

```python
# Creation du repertoire
if "data" not in os.listdir("./"):
     os.mkdir("data")
        
n = 100 # Nombre de points
nt = 10    # Nombre d'essais
t = np.linspace(0., 10., n) # Temps
d = np.linspace(0., 1., n) # Deplacement
for i in range(nt):
    k = np.random.rand()  
    f = k * d**2 + (np.random.rand(n) -.5)* .1
    data = pd.DataFrame({"t":t, "d":d, "f":f}) 
    data.to_csv("data/test_{0}.csv".format(i)) # Sauvegarde csv
    data.to_excel("data/test_{0}.xls".format(i)) # Sauvgarde xls
  
```

```python
from scipy.optimize import curve_fit
path = "data/test_0.csv" # je veux lire ça !
pd.read_csv(path, index_col = 0)

def fonc(x, k):
    """
    Forme théorique de notre essai
    """
    return k * x**2

class Essai(object):
    def __init__(self, path): #Le constructeur 
        self.data = pd.read_csv(path, index_col = 0)
    
    def k(self):
        """
        Finds the value of K
        """
        popt, pcov = curve_fit(fonc, self.data.d, self.data.f)   
        k = popt[0]
        return k

colors = "rgbmyck"    
fig = plt.figure()
for i in range(nt):
    c = colors[i%len(colors)]
    e = Essai("data/test_{0}.csv".format(i))
    plt.plot(e.data.d, e.data.f, "o"+c, markersize = 3)
    plt.plot(e.data.d, fonc(e.data.d, e.k()), "-"+c)
plt.grid()
plt.show()

```

```python
# Le vecteur
class Vecteur(object):
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    def norme(self):
        return (self.x**2 + self.y**2 + self.z**2)**.5
    
    def normaliser(self):
        n = self.norme()
        self.x = self.x / n
        self.y /= n
        self.z /= n
    
    def fixer_norme(self, n):
        k = n / self.n
        self.x *= k
        self.y *= k
        self.z *= k
    
    n = property(norme, fixer_norme)
    
        
        
v = Vecteur(1, 1, 0)   
v.norme()

v.n = 5
v.x,v.y, v.z
```
